```python
values_after_1_epoch = torch.zeros(155)

for it in range(0, 155):
    i = 64*(it+1)
    print(it)
    if i !=320:
        path = f"checkpoints_logits/(5)epoch_1idx_{i}.pt"
    
    
        checkpoint = torch.load(path, map_location=device, weights_only=True)
        loss = checkpoint['loss']
        
        
        values_after_1_epoch[it]=loss
    else:
        values_after_1_epoch[it]=5.1613
        
    



 



```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

steps = np.arange(64, 9984, 64)

sns.set_style("darkgrid")
plt.title("KD in the first epoch")
plt.plot(steps, valuesnew.detach().numpy(), label = "wo 1st phase")
plt.plot(steps,values_after_1_epoch.detach().numpy(), label = "with 1st phase" )
plt.xlabel("iteration")
plt.ylabel("loss value")
plt.legend()
plt.show()
```

```python
kd_vl = torch.empty(3, 20)

for j in range (0, 3):
    for i in range(0, 20):
        if i == 0:
            it = 64
        else:
            it = i*512
        path = f"checkpoints_logits/epoch_{j+4}idx_{it}.pt"
        checkpoint = torch.load(path, map_location=device, weights_only=True)
        loss = checkpoint['loss']
        kd_vl[j][i] = loss
        print(j, i)
```

```python


its = np.arange(0,  10240, 512)
plt.title("Logits alignment with KL divergence")
plt.ylabel("Loss values")
plt.xlabel("Iterations")
plt.plot(its, kd_vl[0][:].detach().numpy(), label = "5th epoch")
plt.plot(its, kd_vl[1][:].detach().numpy(), label = "6th epoch")
plt.plot(its, kd_vl[2][:].detach().numpy(), label ="7th epoch")
plt.legend()
plt.show()
```
